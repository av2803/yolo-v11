```python
import cv2
import torch
from ultralytics import YOLO
import os

# Load the YOLOv11 model
model = YOLO('C:/Users/adity/Downloads/capstone yolo/runs/detect/train3/weights/best.pt')

# Define input and output folders
input_folder = 'C:/Users/adity/Downloads/capstone yolo/PKlot.v2-pklot-new.yolov11/test/images'
output_folder = 'C:/Users/adity/Downloads/capstone yolo/runs/detect/result'
os.makedirs(output_folder, exist_ok=True)

# Define colors for different labels
colors = {0: (255, 0, 0), 1: (0, 255, 0)}  # 0 = Blue for unoccupied, 1 = Green for occupied

# Process each image
for img_name in os.listdir(input_folder):
    img_path = os.path.join(input_folder, img_name)
    img = cv2.imread(img_path)
    
    # Perform detection
    results = model(img)
    
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()
        classes = result.boxes.cls.cpu().numpy()
        
        for box, cls in zip(boxes, classes):
            x1, y1, x2, y2 = map(int, box)
            label = int(cls)
            color = colors.get(label, (0, 255, 255))  # Default color if not defined
            
            # Draw bounding box and label
            cv2.rectangle(img, (x1, y1), (x2, y2), color, 2)
            cv2.putText(img, f'{label}', (x1, y1 - 5), 
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    # Save the output
    output_path = os.path.join(output_folder, img_name)
    cv2.imwrite(output_path, img)

print(f"Processed images saved in '{output_folder}'")
```
